```python
%pylab inline

#load the data
import json
json_data = open('closed.json.txt', 'r')
raw = json.load(json_data)
```

```python
#Question 1
import pandas as pd
test = []

#define function to extract information from the .txt file
def getinfo(boo):
    import numpy as np
    Title = boo['title']
    Created = boo['created_at']
    Labels = boo['labels']
    Closed = boo['closed_at']
    User = boo['user']['login']
    ID = boo['id']
    ans = [Title,Created,Labels,Closed,User,ID]
    test.append(ans)

#run it
for i in range(len(raw)):
    getinfo(raw[i])

#save it as a DataFrame
data = pd.DataFrame(test, columns = ['Title','Created','Labels','Closed','User','ID'])
pd.options.display.line_width = 1000
data.head(10) #data is the DataFrame
```

```python
#Question 2
data2 = data.drop_duplicates(cols='ID')
data2.head(10) #data2 is the DataFrame with duplicate ID rows dropped
```

```python
#Question 3
data2['Created'] = pd.to_datetime(data2['Created'])
data2['Closed'] = pd.to_datetime(data2['Closed'])
data2.dtypes #to show that Created and Closed columns have been changed to datetime type
```

```python
#Question 4

#set Created_at column as index
data2.set_index('Created',inplace=True)
years= data2.index.year
```

```python
#get number of issues per year
annual_count = data2['ID'].groupby(years).count()

#get number of distinct users creating issues per year
annual_nunique = data2['User'].groupby(years).nunique()

#concaternate both
annual_count_and_nunique = pd.concat([annual_count, annual_nunique], 
                               axis=1, keys=['count', 'nunique'])
print annual_count_and_nunique

#now resample to number of issues per month
num_issues_per_month = data2['ID'].resample('M', how='count')

#and plot it
num_issues_per_month.plot(title='Number of issues created per month')
```

```python
#now resample to number of distinct users creating issues per month
num_users_per_month = data2['User'].resample('M',how='nunique')

#and plot it
num_users_per_month.plot(title='Number of distinct users per month')
```

```python
#Question 5
from datetime import datetime

#calculate the deltat and convert to seconds using total_seconds
data3 = data.drop_duplicates(cols='ID')
closed = [datetime.strptime(x, '%Y-%m-%dT%H:%M:%SZ') for x in data3.Closed]
created = [datetime.strptime(y, '%Y-%m-%dT%H:%M:%SZ') for y in data3.Created]
delta = [(closed[z]-created[z]).total_seconds() for z in range(len(closed))]
data3['deltat_seconds'] = delta

data3['Created'] = pd.to_datetime(data3['Created'])
data3['Closed'] = pd.to_datetime(data3['Closed'])
data3.set_index('Created',inplace=True)
years= data3.index.year

#get mean deltat(s) per year
annual_meandelta = data3['deltat_seconds'].groupby(years).mean()
annual_meandelta

#resample to mean deltat(s) per month
mean_delta_per_month = data3['deltat_seconds'].resample('M', how='mean')
mean_delta_per_month

#convert deltat back to days
def to_days(time):
    time = time/86400 #60*60*24
    return time

mean_delta_per_month_days = to_days(mean_delta_per_month)
mean_delta_per_month_days.values[7] = 0. #replace NaN with 0.

#concaternate # issues per month and mean deltat(days) into a table
monthly_count_and_delta = pd.concat([num_issues_per_month, mean_delta_per_month_days], 
                               axis=1, keys=['# issues per month', 'mean deltat(days) to close'])

print monthly_count_and_delta

#and plot it
mean_delta_per_month_days.plot(title='Mean number of days per month for issues to be closed')
```

```python
#Question 6
test2=[]

#define function to extract comments from the .txt file
def getcomments(boo):
    for j in range(len(boo['comments'])):
        Author = boo['comments'][j]['author']
        Created = boo['comments'][j]['created']
        Comments = boo['comments'][j]['text']
        Updated = boo['comments'][j]['updated']
        ID = boo['id']
        ans = [Author,Created,Comments,Updated,ID]
        test2.append(ans)

#run it
for i in range(len(raw)):
    getcomments(raw[i])

#save it as a DataFrame
data4 = pd.DataFrame(test2, columns = ['Author','Created','Comments','Updated','ID'])

#drop duplicate comments
data4 = data4.drop_duplicates(cols='Comments')
data4.head(10)
```

```python
#convert Created column to datetime format
temp = data4.Created.values.astype(int)*1000000
temp = temp.astype(object)
data4.Created = temp
data4['Created'] = pd.to_datetime(data4['Created'])
data4.head(10)
```

```python
#Question 7
#Total number of issue comments
#The "chattiest" user (most number of comments)
#The percentage of total comments made by the chattiest users
#The number of distinct participants in the issue comments

data4.set_index('Created',inplace=True)
years= data4.index.year
```

```python
#get total number of issue comments per month
annual_comments_count = data4['Comments'].groupby(years).count()
num_comments_per_month = data4['Comments'].resample('M', how='count')

#get "chattiest" user per month
annual_user_count = data4['Author'].groupby(years).value_counts()
num_comments_user_per_month = data4['Author'].resample('M', how=['value_counts'])
index1 = [a for a,b in num_comments_user_per_month.index]
num_comments_user_per_month["index1"] = index1 #introduce the index as a separate column in order to use drop_duplicates
chattiest_user_per_month = num_comments_user_per_month.drop_duplicates(cols='index1') #this is to select the most frequent commenter by month
del chattiest_user_per_month["index1"] #then delete the column
chattiest_user=chattiest_user_per_month.index.values 
chattiest_user = [chattiest_user[x][1] for x in range(len(chattiest_user_per_month))]
chattiest_user.insert(7,'none') #add value of 'none' to the newly added row for 'chattiest user'
chattiest_user = array(chattiest_user) #convert to array

#get % of total comments made by the chattiest commenters per month
num_comments_user=chattiest_user_per_month['value_counts'].values
user = num_comments_user.tolist() 
user.insert(7,0) #add value of '0' to the newly added row for '# comments'
num_comments_user = array(user) 
num_comments_all = num_comments_per_month.values
num_comments_all[7] = 1 #changing value from 0 to 1 to avoid division by zero for % calculation
percentage = [float(a) / float(b) *100 for a,b in zip(num_comments_user, num_comments_all)]
percentage = array(percentage)
percentage[7] = NaN #add value of 'NaN' to newly added row for 'percentage'
num_comments_all[7] = 0 #change value back to 0

#get number of unique commenters per month
annual_commenters_count = data4['Author'].groupby(years).nunique()
num_commenters_per_month = data4['Author'].resample('M',how='nunique')
```

```python
#now put everything together in 1 DataFrame
everything = pd.concat([num_comments_per_month, num_commenters_per_month], 
                               axis=1, keys=['total # comments', 'total # commenters'])
everything['chattiest user'] = chattiest_user
everything['% of comments'] = percentage
everything #everything is the DataFrame with total # of comments, total # of commenters, the chattiest commenter and the % of total comments
            #contributed by the chattiest commenter
```

```python
#Question 8
data5 = data2
del data5['Title']
del data5['Closed']
del data5['User']

LI = [] #define empty set

#Create the list of tuples in LI
for i in range(len(data5.index)):
    if len(data5.Labels[i]) == 0:
        LI.append(('None',data5.ID[i]))
    elif len(data5.Labels[i]) != 0:
        for j in range(len(data5.Labels[i])):
            LI.append((data5.Labels[i][j]['color'],data5.ID[i]))
            LI.append((data5.Labels[i][j]['name'],data5.ID[i]))
            LI.append((data5.Labels[i][j]['url'],data5.ID[i]))
            
#Then create the DataFrame
data6 = pd.DataFrame(LI,columns = ['Labels','ID'])
data6.head(10) #data6 is the DataFrame which contains the Labels and ID numbers
```

```python
#Question 9
import pandas as pd
data7 = data.drop_duplicates(cols='ID')
del data7['Labels']
data8 = pd.merge(data7,data6,how='left', on='ID') #merge both DataFrames
data8['Created'] = pd.to_datetime(data8['Created'])
data8['Closed'] = pd.to_datetime(data8['Closed'])
data8['deltat'] = data8['Closed']-data8['Created']

#change deltat to a floating point number in days
temptemp = data8.deltat.values.astype(float)
temptemp = temptemp/(1000000000*60*60*24)
data8['deltat (days)'] = temptemp #add it to the DataFrame
data8.head(10) #data8 is the merged DataFrame with an additional column of # days for issue to close as a floating point number
```

```python
#Question 10
labels = data8.Labels
avg_time_label_type = data8['deltat (days)'].groupby(labels).mean() #calculate mean time to close by label type
print avg_time_label_type

data8.set_index('Created',inplace=True)
```

```python
data9 = data8[data8.Labels == 'Bug']
years9 = data9.index.year
data10 = data8[data8.Labels == 'Enhancement']
years10 = data10.index.year

#get mean number of days for Bug issues to close per year
annual_mean_Bug = data9['deltat (days)'].groupby(years9).mean()

#get mean number of days for Enhancement issues to close per year
annual_mean_Enhancement = data10['deltat (days)'].groupby(years10).mean()

#concaternate both
annual_mean_BE = pd.concat([annual_mean_Bug, annual_mean_Enhancement], 
                               axis=1, keys=['Bug', 'Enhancement'])
print annual_mean_BE

#now resample to mean number of days per month for Bug and Enhancement issues
B_mean_days_per_month = data9['deltat (days)'].resample('M', how='mean')
E_mean_days_per_month = data10['deltat (days)'].resample('M', how='mean')

#and plot it (fill nan's with 0.)
B_mean_days_per_month.fillna(0.).plot(title='Mean number of days taken for Bug/Enhancement issues to close per month',
                                        legend=True,label='Bug issues')
E_mean_days_per_month.fillna(0.).plot(legend=True, label='Enhancement issues')
```
