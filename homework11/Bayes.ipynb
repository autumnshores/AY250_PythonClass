#### 2. The full-season batting average of each player versus the posterior mean from (d) (Include error bars to show the 95% CI).

```python
#calculate lower and upper error limits
low=[]
high=[]
for i in range(len(pos_mean)):
    low.append(pos_mean[i]-lower[i])
for i in range(len(pos_mean)):
    high.append(upper[i]-pos_mean[i])
```

```python
#plot graph
f2, ax2 = plt.subplots()
xaxis = arange(1,14)
ax2.plot(xaxis, AVG2,'bo', label='full season')
ax2.plot(xaxis, pos_mean, 'ro', label='posterior mean')
ax2.errorbar(xaxis, pos_mean,yerr=[low,high])
ax2.set_xlabel('Player')
ax2.set_ylabel('Average')
ax2.legend(loc='best',fontsize='10')
plt.title('Plot of full season batting average versus posterior mean batting average')
```
