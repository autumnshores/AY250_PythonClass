```python
%pylab inline
import numpy as np
import scipy
import scipy.io
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 1: enter global information

```python
patient = "CP5"
elecs_all = range(1,109,1)
bad_elecs = [93,94]
elecs = list(set(elecs_all).difference(set(bad_elecs)))
srate = 1000
ANsrate = 1000
task = "passive"
ictal = [[36,38],[79,82],[118.5,120.5]]
```

## Step 2: load and view the analog speaker file

```python
#data are .mat files that are generated by extracting the raw files from the TDT/Nihon Koden recording system in hospitals
spkrmat = scipy.io.loadmat('spkr.mat')
spkr = spkrmat['spkr'][0]
plt.figure
plt.plot(spkr)
plt.title('speaker file plot')
plt.xlabel('samples')
plt.show()
```

## Step 3: Find auditory events

```python
import find_aud_evs
from find_aud_evs import *

spkr = spkr - mean(spkr)
thresh = 10 #edit threshold manually

#find onset and offset times of the auditory stimuli/events
t = find_aud_evs(spkr,thresh,ANsrate,len(spkr))
```

```python
#show it on the spkr plot
plt.figure
plt.plot(spkr)
plt.title('speaker file plot')
plt.xlabel('samples')
for i in range(len(t)):
    plt.axvline(x=t[i][0], color = 'r')
    plt.axvline(x=t[i][1], color = 'k')
plt.show()
```

## Step 4: load and view the electrode channels

```python
#data are .mat files that are generated by extracting the raw files from the TDT/Nihon Koden recording system in hospitals
gdatmat = scipy.io.loadmat('gdat.mat')
gdat = gdatmat['gdat']
```

```python
#view the EEG channel file, specify channel e (see readme.txt file for further comments on this)
e = 108
plt.figure
plt.plot(gdat[e-1])
plt.title('EEG plot for channel '+str(e))
plt.xlabel('samples')
plt.ylabel('microVolts')
plt.show()
```

## Step 5: Common Average Reference (CAR) - a method for re-referencing electrode signals

```python
import CAR
from CAR import *

Enum = len(elecs) #total number of electrodes
grouping = 27. #calculating the CAR in smaller groups

gdat_CAR = CAR(Enum,grouping,gdat,elecs,bad_elecs)
```

## Step 6: generate Spectrograms

```python
import create_spect
from create_spect import *

compress = 3 #compression factor to compress the data
maxfreq = 300 #upper frequency range (usually up to 250Hz is enough for ECoG)
e = 36 #electrode number we want to generate the spectrogram for

gdat_SPEC = create_spect(gdat_CAR[e],compress,srate,maxfreq)
```

## Step 7: Create events dataframe

```python
#load excel file containing the list of stimulus names played during the experiment, which is obtained from the Eprime presentation software
xl = pd.ExcelFile('CP5_passive_block2.xls')
df = xl.parse("Sheet1", header=None)
df.columns=['stimulus']

#add onset times and offset times calculated in step 5 to dataframe
#also mark the distractor stimuli (meows)
onsets = []
offsets = []
meow = []
for i in range(len(t)):
    onsets.append(t[i][0])
    offsets.append(t[i][1])
    if df['stimulus'][i] == 'meow':
        meow.append(1)
    else: meow.append(0)
df['onsets'] = onsets
df['offsets'] = offsets
df['meow'] = meow
```

```python
#find events that overlap with ictal periods and mark them as bad
ictal = srate*np.array(ictal)
badevent = [0]*len(t)
for j in range(len(t)):
    for k in range(len(ictal)):
        a = arange(round(df['onsets'][j]/float(ANsrate)*srate),round(df['offsets'][j]/float(ANsrate)*srate+1))
        b = arange(round(ictal[k][0]),round(ictal[k][1])+1)
        if len(set(a).intersection(b)) != 0:
            badevent[j] = 1
df['badevent'] = badevent
```
