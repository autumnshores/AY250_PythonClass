```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
from __future__ import division

class ImageFeatures:
    def __init__(self, category, imnum):
        self.category = category
        self.number = imnum
        self.image = plt.imread('50_categories/'+self.category+'/'+self.category+'_'+self.number+'.jpg')
        red, green, blue = [self.image[:,:,i] for i in range(3)]
        self.red = red
        self.green = green
        self.blue = blue
        lumi = .299*self.red + .587*self.green + 0.114*self.blue
        self.gray = lumi
    
    #load the image
    def loadimage(self):
        return plt.imshow(self.image)
    
    #Feature 1: get image size
    def getimsize(self):
        print 'Feature 1: The size of the loaded image is', self.image.shape[0], 'by', self.image.shape[1]
        return self.image.shape[0], self.image.shape[1]
    
    #Feature 2: get mean of red channel colors    
    def getRmean(self):
        Rmean = np.mean(self.red)
        print 'Feature 2: The average of the red channel colors is', Rmean
        self.Rmean = Rmean
        return Rmean
    
    #Feature 3: get mean of green channel colors
    def getGmean(self):
        Gmean = np.mean(self.green)
        print 'Feature 3: The average of the green channel colors is', Gmean
        self.Gmean = Gmean
        return Gmean
     
    #Feature 4: get mean of blue channel colors
    def getBmean(self):
        Bmean = np.mean(self.blue)
        print 'Feature 4: The average of the blue channel colors is', Bmean
        self.Bmean = Bmean
        return Bmean
    
    #Feature 5: get mean of grayscale (converted RGB) colors
    def getGrayscalemean(self):
        Graymean = np.mean(self.gray)
        print 'Feature 5: The average of the grayscale (converted RGB) colors is', Graymean
        self.Graymean = Graymean
        return Graymean
    
    #Feature 6: get mean of cross-correlation between the Red and Green channels
    def getXCorrmeanRG(self):
        from scipy import signal

        XCorrRG = scipy.signal.correlate2d(self.red,self.green,mode='valid')
        #XCorrmeanRG = np.mean(XCorrRG)
        print 'Feature 6: The cross-correlation between the Red and Green channels ("valid" mode) is', XCorrRG
        self.XCorrRG = XCorrRG
        return XCorrRG
    
    #Feature 7: get mean of cross-correlation between the Red and Blue channels
    def getXCorrmeanRB(self):

        XCorrRB = scipy.signal.correlate2d(self.red,self.blue,mode='valid')
        #XCorrmeanRB = np.mean(XCorrRB)
        print 'Feature 7: The cross-correlation between the Red and Blue channels ("valid" mode) is', XCorrRB
        self.XCorrRB = XCorrRB
        return XCorrRB
    
    #Feature 8: get mean of cross-correlation between the Green and Blue channels
    def getXCorrmeanGB(self):

        XCorrGB = scipy.signal.correlate2d(self.green,self.blue,mode='valid')
        #XCorrmeanGB = np.mean(XCorrGB)
        print 'Feature 8: The cross-correlation between the Green and Blue channels ("valid" mode) is', XCorrGB
        self.XCorrGB = XCorrGB
        return XCorrGB
    
    #Feature 9: get proportion of pixels that have entropy above a certain specified threshold
    def getProportionEntropy(self,thresh):
        from skimage.filter.rank import entropy
        from skimage.morphology import disk
        from skimage.util import img_as_ubyte
        #convert grayscale image to uint8
        convert = self.gray.astype(uint8)

        # defining a 8-bit test images
        a8 = img_as_ubyte(convert)

        ent8 = entropy(a8, disk(5)) # pixel value contain 10x the local entropy
        
        tot = ent8.shape[0]*ent8.shape[1]
        prop = (ent8 > thresh).sum()
        rat = prop/tot
        print 'Feature 9: The proportion of pixels in the image that have entropy above a threshold', thresh, 'is', rat
        self.entropy = rat
        return rat
    
    #Feature 10: get proportion of pixels that are part of horizontal edges above a certain specified threshold
    def getProportionHSobelEdges(self,thres):
        from skimage.filter import hsobel

        h_edge_sobel = hsobel(self.gray)
        total = h_edge_sobel.shape[0]*h_edge_sobel.shape[1]
        proportion = (h_edge_sobel > thres).sum()
        ratio = proportion/total
        print 'Feature 10: The proportion of pixels in the image that are part of horizontal edges above a threshold', thres, 'is', ratio
        self.hsobel = ratio
        return ratio
    
    #Feature 11: get proportion of pixels that are part of vertical edges above a certain specified threshold
    def getProportionVSobelEdges(self,thres2):
        from skimage.filter import vsobel

        v_edge_sobel = vsobel(self.gray)
        total2 = v_edge_sobel.shape[0]*v_edge_sobel.shape[1]
        proportion2 = (v_edge_sobel > thres2).sum()
        ratio2 = proportion2/total2
        print 'Feature 11: The proportion of pixels in the image that are part of vertical edges above a threshold', thres2, 'is', ratio2
        self.vsobel = ratio2
        return ratio2
    
    #Feature 12: get proportion of pixels that are part of all edges above a certain specified threshold
    def getProportionSobelEdges(self,thres3):
        from skimage.filter import sobel

        edge_sobel = sobel(self.gray)
        total3 = edge_sobel.shape[0]*edge_sobel.shape[1]
        proportion3 = (edge_sobel > thres3).sum()
        ratio3 = proportion3/total3
        print 'Feature 12: The proportion of pixels in the image that are part of all edges above a threshold', thres3, 'is', ratio3
        self.sobel = ratio3
        return ratio3
    
    #Feature 13: get number of objects after image segmentation
    def getnumObjects(self):
        from scipy import ndimage
        from skimage.morphology import watershed
        from skimage.feature import peak_local_max

        distance = ndimage.distance_transform_edt(self.gray)
        local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((3, 3)),labels=self.gray)
        markers = ndimage.label(local_maxi)[0]
        labels = watershed(-distance, markers, mask=self.gray)
        print 'Feature 13: The number of objects after image segmentation is', labels.max()
        self.numobjects = labels.max()
        return labels.max()
    
    #Feature 14: get number of corners
    def getnumCorners(self):
        from skimage.feature import corner_harris, corner_subpix, corner_peaks
        coords = corner_peaks(corner_harris(self.gray), min_distance=5)
        coords_subpix = corner_subpix(self.gray, coords, window_size=13)
        print 'Feature 14: The number of corners in the image is', len(coords)
        self.numcorners = len(coords)
        return len(coords)
    
    #Feature 15: get number of contours at a certain specified value
    def getnumContours(self,value):
        from skimage import measure

        # Find contours at a constant value
        contours = measure.find_contours(self.gray, value)
        
        print 'Feature 15: The number of contours in the image at a value of', value, 'is', len(contours)
        self.numcontours = len(contours)
        return len(contours)
    
    def SaveAllFeatures(self):
        all = [self.image.shape[0], self.image.shape[1], self.Rmean, self.Gmean, self.Bmean, self.Graymean, int(self.XCorrRG), int(self.XCorrRB), int(self.XCorrGB), self.entropy, self.hsobel, self.vsobel, self.sobel, self.numobjects, self.numcorners, self.numcontours]
        return all
```
